```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
tw22 = pd.read_csv('2022.csv')
tw21 = pd.read_csv('2021.csv')
tw20 = pd.read_csv('2020.csv')
tw19 = pd.read_csv('2019.csv')
tw18 = pd.read_csv('2018.csv')
tw17 = pd.read_csv('2017.csv')
tw16 = pd.read_csv('2016.csv')
tw15 = pd.read_csv('2015.csv')
```

I used 2015 as the base dataset, and all other sets were compared against 2015 to find the common countries. The rest of the code is to rename the columns in a standard way, and to drop unneccesary columns.

```python
tw15.head()
```

```python
tw15.rename(columns = {'Happiness Rank': 'RANK', 'Happiness Score': 'Happiness score', 'Economy (GDP per Capita)': 'Explained by: GDP per capita', 'Family': 'Explained by: Social support', 'Health (Life Expectancy)': 'Explained by: Healthy life expectancy', 'Freedom': 'Explained by: Freedom to make life choices', 'Generosity': 'Explained by: Generosity', 'Trust (Government Corruption)': 'Explained by: Perceptions of corruption', 'Dystopia Residual': 'Dystopia + residual', 'Region': 'Regional indicator'}, inplace = True)
tw15['Country'] = tw15['Country'].astype(str).str.replace('*','')
tw15 = tw15[['RANK', 'Country', 'Regional indicator', 'Happiness score', 'Dystopia + residual', 'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption']]
tw15.head()
```

```python
tw15.shape
```

```python
colTokeep = ['RANK', 'Country', 'Regional indicator', 'Happiness score', 'Dystopia score', 'Dystopia + residual', 'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption']
```

```python
tw22.insert(5, 'Dystopia score', 1.83)
tw22.rename(columns={'Dystopia (1.83) + residual': 'Dystopia + residual'}, inplace = True)
tw22['Country'] = tw22['Country'].str.replace('*','')
tw22["Country"].replace(to_replace = {'Czechia': "Czech Republic", 'Taiwan Province of China': 'Taiwan', 'Hong Kong S.A.R. of China': 'Hong Kong', 'Congo': 'Congo (Kinshasa)'}, inplace = True)
for column in tw22:
    if column not in ["RANK", "Country", "Dystopia score"]:
        tw22[column] = tw22[column].astype(str).str.replace(',', '.').astype(float)
tw22['RANK'] = tw22['RANK'].astype(int)
tw22.dtypes
```

```python
tw22 = tw22.merge(tw15[['Country', 'Regional indicator']], how = 'inner', on ='Country')
tw22.head()
```

```python
tw21.insert(0, "Rank", tw21.index+1)
tw21.rename(columns = {"Rank": 'RANK', 'Country name': 'Country', 'Ladder score': "Happiness score", "Explained by: Log GDP per capita":'Explained by: GDP per capita', 'Ladder score in Dystopia': 'Dystopia score'}, inplace = True)
tw21 = tw21[colTokeep]
tw21['Country'] = tw21['Country'].astype(str).str.replace('*','')
tw21.head()
```

```python
tw20.insert(0, 'RANK', tw20.index + 1)
tw20.rename(columns = {'Country name': 'Country', "Ladder score": 'Happiness score', 'Ladder score in Dystopia': 'Dystopia score', 'Explained by: Log GDP per capita': 'Explained by: GDP per capita'}, inplace = True)
tw20 = tw20[colTokeep]
tw20['Country'] = tw20['Country'].astype(str).str.replace('*','')
tw20.head()
```

```python
tw19.rename(columns = {'Overall rank': 'RANK', 'Country or region': 'Country', 'Score': 'Happiness score', 'GDP per capita': 'Explained by: GDP per capita', 'Social support': 'Explained by: Social support', 'Healthy life expectancy': 'Explained by: Healthy life expectancy', 'Freedom to make life choices': 'Explained by: Freedom to make life choices', 'Generosity': 'Explained by: Generosity', 'Perceptions of corruption': 'Explained by: Perceptions of corruption'}, inplace = True)
tw19.insert(3, 'Dystopia + residual', tw19['Happiness score'] - tw19['Explained by: GDP per capita'] - tw19['Explained by: Social support'] - tw19['Explained by: Healthy life expectancy'] - tw19['Explained by: Freedom to make life choices'] - tw19['Explained by: Generosity'] - tw19['Explained by: Perceptions of corruption'])
tw19['Country'] = tw19['Country'].astype(str).str.replace('*','')
tw19.replace(to_replace = {'Trinidad & Tobago': 'Trinidad and Tobago', 'Northern Cyprus': 'North Cyprus', 'North Macedonia': 'Macedonia','Somalia': 'Somaliland region', 'South Sudan': 'Sudan'}, inplace = True)
tw19 = tw19.merge(tw15[['Country', 'Regional indicator']], how = 'inner', on ='Country')
tw19 = tw19[['RANK', 'Country', 'Regional indicator', 'Happiness score', 'Dystopia + residual', 'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption']]
tw19.head()
```

```python
tw18.rename(columns = {'Overall rank': 'RANK', 'Country or region': 'Country', 'Score': 'Happiness score', 'GDP per capita': 'Explained by: GDP per capita', 'Social support': 'Explained by: Social support', 'Healthy life expectancy': 'Explained by: Healthy life expectancy', 'Freedom to make life choices': 'Explained by: Freedom to make life choices', 'Generosity': 'Explained by: Generosity', 'Perceptions of corruption': 'Explained by: Perceptions of corruption'}, inplace = True)
tw18.insert(3, 'Dystopia + residual', tw18['Happiness score'] - tw18['Explained by: GDP per capita'] - tw18['Explained by: Social support'] - tw18['Explained by: Healthy life expectancy'] - tw18['Explained by: Freedom to make life choices'] - tw18['Explained by: Generosity'] - tw18['Explained by: Perceptions of corruption'])
tw18['Country'] = tw18['Country'].astype(str).str.replace('*','')
tw18.replace(to_replace = {'Trinidad & Tobago': 'Trinidad and Tobago', 'Northern Cyprus': 'North Cyprus', 'Somalia': 'Somaliland region', 'South Sudan': 'Sudan'}, inplace = True)
tw18 = tw18.merge(tw15[['Country', 'Regional indicator']], how = 'inner', on ='Country')
tw18 = tw18[['RANK', 'Country', 'Regional indicator', 'Happiness score', 'Dystopia + residual', 'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption']]
tw18.head()
```

```python
tw17.rename(columns = {'Happiness.Rank': 'RANK', 'Happiness.Score': 'Happiness score', 'Economy..GDP.per.Capita.': 'Explained by: GDP per capita', 'Family': 'Explained by: Social support', 'Health..Life.Expectancy.': 'Explained by: Healthy life expectancy', 'Freedom': 'Explained by: Freedom to make life choices', 'Generosity': 'Explained by: Generosity', 'Trust..Government.Corruption.': 'Explained by: Perceptions of corruption', 'Dystopia.Residual': 'Dystopia + residual'}, inplace = True)
tw17['Country'] = tw17['Country'].astype(str).str.replace('*','')
tw17.replace(to_replace = {'Taiwan Province of China': 'Taiwan', 'Hong Kong S.A.R., China': 'Hong Kong', 'Somalia': 'Somaliland region', 'South Sudan': 'Sudan'}, inplace = True)
tw17 = tw17.merge(tw15[['Country', 'Regional indicator']], how = 'inner', on ='Country')
tw17 = tw17[['RANK', 'Country', 'Regional indicator', 'Happiness score', 'Dystopia + residual', 'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption']]
tw17.head()
```
